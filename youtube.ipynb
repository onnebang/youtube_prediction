# 유튜브 인기동영상 분석

```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
data = pd.read_csv('FOOD.csv', engine='python', encoding='UTF8')
data.head()
```

```python
model = smf.ols(formula = 'views ~ likes', data = data)
result = model.fit()
result.summary()
```

```python
model = smf.ols(formula = 'views ~ dislikes', data = data)
result = model.fit()
result.summary()
```

```python
model = smf.ols(formula = 'views ~ comments', data = data)
result = model.fit()
result.summary()
```

```python
model = smf.ols(formula = 'views ~ likes+dislikes', data = data)
result = model.fit()
result.summary()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas
from statsmodels.formula.api import ols
res = ols('views ~ likes', data=data).fit()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fitted = res.predict(data)
residual = data['views'] - fitted

sns.regplot(fitted, residual, lowess=True, line_kws={'color': 'red'})
plt.plot([fitted.min(), fitted.max()], [0, 0], '--', color='grey')
```

```python
import statsmodels.api as sm
others = list(set(data.columns).difference(set(["views", "likes"])))
p, resids = sm.graphics.plot_partregress("views", "likes", others, data = data, ret_coords = True)
plt.show()
```
